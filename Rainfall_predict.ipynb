```python
import pickle
import pandas as pd
```

```python
#load the trained model & feature names from the pickle file
with open("model_RF.pkl","rb") as file:
    model_data= pickle.load(file)
```

```python
model = model_data['model']
feature_names = model_data["feature names"]
```

```python
input_data = (1020.5,20.4,98,80,0.0,41.0,13.7)
input_df = pd.DataFrame([input_data],columns=feature_names)
```

```python
# Make prediction
prediction = model.predict(input_df)

# Print the result
print("Predicted value:", "rainfall" if  prediction[0]==1 else "no rainfall" )
```
