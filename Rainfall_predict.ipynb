```python
import pickle
import pandas as pd

```

```python
with open("model_RF.pkl", "rb") as file:
    loaded_model = pickle.load(file)

# Extract the model, feature names, and scaler
model = loaded_model['model']
feature_names = loaded_model["feature_names"]
scaler = loaded_model['scaler']

print("Model expects features:", feature_names)
print("Number of features:", len(feature_names))

```

```python
# Example input data (replace with real values)
input_data = (1015.9, 19.9, 95, 81, 40.0, 13.7)  

# Convert to DataFrame with correct column names
input_df = pd.DataFrame([input_data], columns=feature_names)



```

```python
prediction = model.predict(input_df)

# Print the result
print("Prediction Result:", "Rainfall" if prediction[0] == 1 else "No Rainfall")
```
